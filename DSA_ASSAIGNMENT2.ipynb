# 3.Plot a bar graph of male vs female buyers participated in the sales.


```python
carf_female=carf.drop(carf[carf["Buyer Gender"]=="Male"].index)
```

```python
carf_male=carf.drop(carf[carf["Buyer Gender"]=="Female"].index)
```

```python
male=carf_male.index
number_of_males=len(male)
female=carf_female.index
number_of_females=len(female)
```

```python
df_1={"Buyer Gender":['Male','Female'],"count":[number_of_males,number_of_females]}
df=pd.DataFrame(df_1)
c=['orange','blue']
df.plot(kind='bar',rot='0', x='Buyer Gender',y='count',color=c,legend=None)
plt.title('Male vs Female buyers participated in the sales')
plt.xlabel('Gender')
plt.ylabel('count')
```
